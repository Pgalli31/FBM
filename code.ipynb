```python
from fbm import FBM
import matplotlib.pyplot as plt
import numpy as np
```

### Hosking

```python
f = FBM(n=1000, hurst=0.2, length=1, method="hosking")

fbm_samples = f.fbm()
t_values = f.times()

plt.plot(t_values, fbm_samples)
plt.title("Fractional Brownian Motion (Hosking H = 0.2)")
plt.xlabel("Time")
plt.ylabel("FBM Value")
plt.show()
```

```python
f = FBM(n=1000, hurst=0.2, length=1, method="hosking")

fgn_samples = f.fgn()
t_values = np.linspace(0, 1, 1000)

plt.plot(t_values, fgn_samples)
plt.title("Fractional Gaussian Noise (Hosking H = 0.2)")
plt.xlabel("Time")
plt.ylabel("FGN Value")
plt.show()
```

```python
import pandas as pd
```

```python
data = pd.read_csv("AAPL_data.csv")
```

```python
data
```

```python
data['date'] = pd.to_datetime(data['date'], utc=True)
```

```python
data = data.set_index(data['date']).sort_index()
```

```python
data.head(5)
```

```python
data['close'].plot(figsize=(15,5))
```

```python
close_prices = data['close'].values
```

```python
n = len(close_prices - 1)
f = FBM(n=n, hurst=0.6, length=1, method='daviesharte')
fbm_samples = f.fbm()
fbm_scaled = (fbm_samples - np.mean(fbm_samples)) / np.std(fbm_samples) * np.std(close_prices) + np.mean(close_prices)
t_real = np.linspace(0, 1, n)
t_sim = f.times()

plt.figure(figsize=(10,5))
plt.plot(t_real, close_prices, label="Real Close Prices")
plt.plot(t_sim, fbm_scaled, label="Simulated fBm")
plt.legend()
plt.title("Real vs. Simulated")
plt.show()
```

```python
price_diff = np.diff(close_prices)
log_returns = np.diff(np.log(close_prices))
```

```python
n = len(log_returns)
f = FBM(n=n, hurst=0.6, length=1, method='daviesharte')
fgn_samples = f.fgn()
fgn_scaled = (fgn_samples - np.mean(fgn_samples)) / np.std(fgn_samples) * np.std(log_returns) + np.mean(log_returns)
t_val = np.linspace(0, 1, n)

plt.figure(figsize=(10,5))
plt.plot(t_val, log_returns, label="Real Log Returns")
plt.plot(t_val, fgn_scaled, label="Simulated fGn")
plt.legend()
plt.title("Real vs. Simulated")
plt.show()
```
